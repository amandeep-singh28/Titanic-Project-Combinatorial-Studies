```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r'Datasets\Titanic-Dataset.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.drop(columns = ['Cabin'], inplace = True) # Too many missing values
```

```python
# Dropping Unnecessary Columns
df.drop(columns=['PassengerId', 'Name', 'Ticket'], inplace=True)
```

```python
df['Sex'].unique() # Contains Categorical Values
```

```python
df['Sex'] = df['Sex'].map({'male' : 0, 'female' : 1})
```

```python
df['Sex'].unique() # Now Numerical Values
```

```python
df['Embarked'].unique()
```

```python
df['Embarked'] = df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
```

```python
df['Embarked'].unique()
```

#### Feature Engineering

```python
df.head()
```

```python
# SibSp -> Number of siblings + spouses traveling with the passenger.
# Parch -> Number of parents + children traveling with the passenger.
# We can combine these two columns as FamilySize
df['Family Size'] = df['SibSp'] + df['Parch']
```

```python
df.head()
```

```python
df['IsAlone'] = np.where(df['Family Size'] == 0, 1, 0)
```

```python
df.head()
```

```python
df.to_csv('Data_Cleaning.csv', index = False)
```
